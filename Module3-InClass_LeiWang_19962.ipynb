First, analyze if it's An integer x, then append into the result list with this new score of x.
if it's '+', then plus the sum of the previous two scores.
if it's 'D', then pop the last value and double the value then append to the list.
if it's 'C', then pop the last value and removing it from the record.
Return the sum of all the scores on the record after traverse all the operations.

```python
class Solution:
    def calPoints(self, operations: List[str]) -> int:
        res = []
        for op in operations:
            if op.lstrip('-').isdigit():
                res.append(int(op))
            elif op == '+':
                pre1 = res[-1]
                pre2 = res[-2]
                res.append(pre1+pre2)
            elif op == 'D':
                pre1 = res[-1]
                res.append(pre1*2)
            elif op == 'C':
                res.pop()
        return sum(res)
```

First create a key-value dictionary to store the different types of parentheses, then create a stack to store the temp parentheses from loading the string. loop each of the characters in the str. if the val in the dictionary, then we should append it into the stack. if val is in the keys, then we need to find the another pair of the parenthese which maps to this current type exactly. If now, the length of the stack is 0 or the mapping relathionship does not exist, then we return false, or else, continue to the next loop. At the end of the loop, we check if the length of the stack is 0, then it is valid or else, it is invalid.

```python
class Solution:
    def isValid(self, s: str) -> bool:
        pd = {')': '(', '}': '{', ']': '['}  
        stack = []
        
        for i in s:  #
            if i in pd.values():  
                stack.append(i)
            elif i in pd.keys():  
                if len(stack) == 0 or pd[i] != stack.pop(): 
                    return False
            else:
                return False  
            
        return len(stack) == 0  
```


I implemented a singly linked list with a Node class that holds a value and a pointer to the next node. I manage the list's length and provide methods for getting a value, adding at the head or tail, inserting at a specific index, and deleting a node based on its index.

```python
class Node:
    def __init__(self,val):
        self.val=val
        self.next=None


class MyLinkedList:

    def __init__(self):
        self.head=None
        self.length=0
        

    def get(self, index: int) -> int:
        if index>=self.length:
            return -1
        curr=self.head
        while index:
            curr=curr.next
            index-=1
        return curr.val
        

    def addAtHead(self, val: int) -> None:
        newnode=Node(val)
        if self.head==None:
            self.head=newnode
        else:
            newnode.next=self.head
            self.head=newnode
        self.length+=1
        

    def addAtTail(self, val: int) -> None:
        newnode=Node(val)
        if self.head==None:
            self.head=newnode
        else:
            curr=self.head
            while curr.next!=None:
                curr=curr.next
            curr.next=newnode
        self.length+=1


    def addAtIndex(self, index: int, val: int) -> None:
        if index < 0 or index > self.length:
            return
        elif index == 0:
           self.addAtHead(val)
        elif index == self.length:
           self.addAtTail(val)
        else:
            new_node = Node(val)
            curr = self.head
            for _ in range(index - 1):
                curr = curr.next
            new_node.next = curr.next
            curr.next = new_node
            self.length += 1


    def deleteAtIndex(self, index: int) -> None:
        if index < 0 or index >= self.length:
            return
        elif index == 0:
           self.head = self.head.next
        elif index == self.length - 1:
            curr = self.head
            while curr.next.next is not None:
                curr = curr.next
            curr.next = None
        else:
            curr = self.head
            i = index - 1
            while i:
                curr = curr.next
                i -= 1
            curr.next = curr.next.next
        self.length -= 1
```

 I used two queues by pushing elements into one queue to implement the Stack. For pop and top operations, I transferred all but the last element to the second queue, retrieved the last element, and then restored the first queue to maintain order, ensuring LIFO behavior.

```python
from collections import deque

class MyStack:

    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x: int) -> None:
        self.q1.append(x)

    def pop(self) -> int:
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())
        res = self.q1.popleft()
        self.q1, self.q2 = self.q2, self.q1
        return res

    def top(self) -> int:
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())
        res = self.q1.popleft()
        self.q2.append(res)
        self.q1, self.q2 = self.q2, self.q1
        return res

    def empty(self) -> bool:
        return not self.q1
```

I implemented a recursive solution with memorizing the calculated numbers of distinct ways to climb stairs. I defined a helper function that returns 1 for one step and 2 for two steps, caching results to avoid redundant calculations, thus optimizing performance for larger inputs.

```python
class Solution:
    def climbStairs(self, n: int) -> int:
        memo = [-1] * (n + 1)
        
        def recur(n):
            if n == 1:
                return 1
            if n == 2:
                return 2
            if memo[n] != -1:  
                return memo[n]
            memo[n] = recur(n-1) + recur(n-2)  
            return memo[n]
        
        return recur(n)
```

I implemented a recursive solution to calculate Fibonacci numbers by defining a helper function that directly applies the Fibonacci definition. The function returns 0 for `F(0)` and 1 for `F(1)`, recursively summing the two preceding Fibonacci numbers for larger values of `n`.

```python
class Solution:
    def fib(self, n: int) -> int:
        def recur(n):
            if n == 0:
                return 0
            elif n == 1:
                return 1
            else:
                return recur(n-1) + recur(n-2)
        return recur(n)
        
```
